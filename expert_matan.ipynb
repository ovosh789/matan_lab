**Задание 1.** следует ли асимптотическая устойчивость точки $x^*$ из условия


$$
\exists \delta_0 > 0:\quad |x_0 - x^*| < \delta_0 \ \Longrightarrow\ x_n \xrightarrow[n\to\infty]{} x^* \, ?
$$

**Ответ: нет, не обязательно.**

Асимптотическая устойчивость по определению требует **двух** свойств:

1. **устойчивости по Ляпунову**:
   $$
   \forall \varepsilon > 0\ \exists \delta > 0:\ 
   |x_0-x^*|<\delta \ \Longrightarrow\ |x_n-x^*|<\varepsilon \ \forall n;
   $$
2. **локальной сходимости** $x_n \to x^*$ при достаточно малых $|x_0-x^*|$.

Данное в условии требование обеспечивает только пункт (2), но **не гарантирует** пункт (1):
траектория может в процессе движения выходить из сколь угодно малой окрестности $x^*$,
а затем всё равно вернуться и сойтись к $x^*$.

**Контрпример.** Рассмотрим отображение $f:\mathbb{R}\to\mathbb{R}$ и неподвижную точку $x^*=0$:
$$
f(x)=
\begin{cases}
0, & x=0 \ \text{или}\ x=1,\\[4pt]
1, & 0<|x|<1,\\[4pt]
0, & |x|\ge 1.
\end{cases}
$$

Точка $0$ неподвижна, так как $f(0)=0$.

- Если $|x_0|<1$ и $x_0\ne 0$, то $x_1=1$, а затем $x_2=f(1)=0$ и далее $x_n=0$.
  Поэтому при $\delta_0=1$ выполнено:
  $$
  |x_0-0|<\delta_0 \ \Longrightarrow\ x_n \to 0.
  $$

- Однако устойчивости по Ляпунову нет. Возьмём $\varepsilon=\dfrac12$.
  Какое бы $\delta>0$ ни выбрали, можно взять $x_0$ так, что
  $$
  0<|x_0|<\min(\delta,1).
  $$
  Тогда $x_1=1$, следовательно
  $$
  |x_1-0|=1>\varepsilon,
  $$
  то есть траектория выходит из $\varepsilon$-окрестности нуля уже на первом шаге.

Итак, существует отображение, для которого условие
$$
|x_0-x^*|<\delta_0 \Rightarrow x_n\to x^*
$$
выполнено, но устойчивость по Ляпунову нарушается. Следовательно, из одной лишь сходимости
**асимптотическая устойчивость не следует**.

**Задание 2.**  При $0<r<1$ неподвижная точка $x^*=0$ устойчива. Является ли она асимптотически устойчивой?

Рассматриваем логистическое отображение
$$
x_{n+1}=r x_n(1-x_n), \qquad 0<r<1,
$$
и начальные условия $x_0\in(0,1)$ (как в начале лабы).

Очевидно, $x^*=0$ — неподвижная точка, так как $f(0)=0$.


## 1) Устойчивость

Нужно доказать:
$$
\forall \varepsilon>0\ \exists \delta>0:\ |x_0-0|<\delta \Rightarrow |x_n-0|<\varepsilon\ \forall n.
$$

Возьмём
$$
\delta=\min\{\varepsilon,\,1\}.
$$
Пусть $0<x_0<\delta\le 1$. Тогда по уже доказанной инвариантности при $0<r\le 1$ имеем
$$
0<x_n<1 \quad \forall n.
$$
Отсюда $0<1-x_n<1$, значит
$$
0<x_{n+1}=r x_n(1-x_n)<r x_n<x_n \quad \forall n,
$$
то есть $\{x_n\}$ убывает, и поэтому
$$
0<x_n\le x_0<\delta\le \varepsilon \quad \forall n.
$$
Следовательно, $|x_n-0|<\varepsilon$ для всех $n$, значит точка $x^*=0$ **устойчива**.


## 2) Асимптотическая устойчивость

По определению нужно дополнительно: $\exists\delta_0>0$, что из $|x_0-0|<\delta_0$ следует $x_n\to 0$.

Возьмём $\delta_0=1$. Тогда при $0<x_0<1$ мы уже получили:
$$
0<x_{n+1}<x_n \quad \forall n,
$$
значит $\{x_n\}$ монотонно убывает и ограничена снизу нулём, поэтому существует предел
$$
L=\lim_{n\to\infty}x_n \ge 0.
$$

Переходим к пределу в рекуррентной формуле:
$$
L=rL(1-L).
$$
Отсюда
$$
L=0 \quad \text{или} \quad L=1-\frac1r.
$$
Но при $0<r<1$ величина $1-\frac1r<0$, а $L\ge 0$, значит остаётся только
$$
L=0.
$$


## Вывод

При $0<r<1$ и $x_0\in(0,1)$ неподвижная точка $x^*=0$ является **устойчивой** и
**асимптотически устойчивой**, так как $x_n\to 0$.

**Задание 3.**  Докажите, что точка $x^*=0$ при $r\in(2,3)$ является неустойчивой

Рассматриваем логистическое отображение
$$
x_{n+1}=f(x_n)=r\,x_n(1-x_n), \qquad r\in(2,3).
$$
Точка $x^*=0$ неподвижна, так как $f(0)=0$.

По определению, $x^*=0$ **устойчива**, если
$$
\forall \varepsilon>0\ \exists \delta>0:\ |x_0|<\delta \ \Longrightarrow\ |x_n|<\varepsilon\ \forall n.
$$
Докажем, что это неверно (то есть $0$ неустойчива).


## Шаг 1. Оценка роста вблизи нуля

Пусть $0<x\le \dfrac12$. Тогда $1-x\ge\dfrac12$, поэтому
$$
f(x)=r x(1-x)\ \ge\ r x\cdot \frac12\ =\ \frac{r}{2}\,x.
$$
Обозначим
$$
q=\frac{r}{2}.
$$
Так как $r\in(2,3)$, то $q>1$.


## Шаг 2. Неустойчивость по определению

Зафиксируем
$$
\varepsilon=\frac12.
$$
Покажем: для любого $\delta>0$ найдётся $x_0\in(0,\delta)$ и номер $n$, что $x_n\ge \varepsilon$.

Пусть дано произвольное $\delta>0$. Выберем целое $m$ настолько большим, что
$$
x_0=\frac12\,q^{-m} < \delta.
$$
(Это возможно, так как $q>1$ и $q^{-m}\to 0$ при $m\to\infty$.)

Рассмотрим траекторию $\{x_n\}$, стартующую из этого $x_0$.

- Если уже для некоторого $k\le m$ выполнено $x_k\ge \dfrac12$, то мы получили
  $|x_k-0| \ge \varepsilon$ и требуемое найдено.

- Иначе предположим, что
  $$
  x_k<\frac12 \quad \text{для всех } k=0,1,\dots,m.
  $$
  Тогда к каждому из этих $x_k$ применима оценка из шага 1:
  $$
  x_{k+1} \ge q\,x_k \quad (k=0,1,\dots,m-1).
  $$
  По индукции получаем
  $$
  x_m \ge q^m x_0 = q^m\cdot\frac12\,q^{-m}=\frac12,
  $$
  что противоречит предположению $x_m<\dfrac12$.

Значит, обязательно существует $k\le m$, для которого $x_k\ge \dfrac12=\varepsilon$.


## Вывод

Мы нашли $\varepsilon=\dfrac12$ такое, что **для любого** $\delta>0$ можно выбрать $0<x_0<\delta$,
при котором некоторая итерация выходит за пределы $\varepsilon$-окрестности нуля.
Следовательно, неподвижная точка $x^*=0$ при $r\in(2,3)$ является **неустойчивой**.

**Задание 4** Напишите функцию, которая для заданных $x_0$, $y_0=x_0+\varepsilon$, $r$ строит две траектории лестницы Ламерея на одном графике. Постройте этот график для $r=4$. Как можно интерпретировать результаты?

Это демонстрирует чувствительность к начальным условиям - ключевое свойство хаотических систем, также известное как эффект бабочки.


```python
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'

def f(x, r):
    return r * x * (1 - x)

def cobweb_path(x0, r, n_steps):
    """Возвращает ломаную для лестницы Ламерея:"""
    xs = [x0]
    x = x0
    for _ in range(n_steps):
        x = f(x, r)
        xs.append(x)

    path_x = [xs[0]]
    path_y = [0.0]
    for k in range(len(xs) - 1):
        xk = xs[k]
        xk1 = xs[k+1]

        path_x.append(xk)
        path_y.append(xk1)

        path_x.append(xk1)
        path_y.append(xk1)
    return xs, path_x, path_y

r = 4.0
x0 = 0.2000
eps = 1e-6
y0 = x0 + eps
n_steps = 40

grid = np.linspace(0, 1, 600)
fg = f(grid, r)

xs, px, py = cobweb_path(x0, r, n_steps)
ys, qx, qy = cobweb_path(y0, r, n_steps)

plt.figure(figsize=(8.5, 8.5))

plt.plot(grid, fg, linewidth=2, label=r'$y=f(x)=rx(1-x)$')
plt.plot(grid, grid, '--', alpha=0.6, label=r'$y=x$')

plt.plot(px, py, marker='.', linewidth=1.4, alpha=0.75, label=fr'траектория $x_0={x0}$')
plt.plot(qx, qy, marker='.', linewidth=1.4, alpha=0.75, label=fr'траектория $y_0=x_0+\varepsilon$')

plt.title(fr'Лестницы Ламерея для двух близких стартов при $r={r}$')
plt.xlabel(r'$x_n$')
plt.ylabel(r'$x_{n+1}$')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.show()
```

- Дальше видно ключевой эффект: **лестницы начинают расходиться**.

- При $r=4$ такое расхождение происходит быстро: малое отличие в начальных условиях
  спустя несколько итераций превращается в заметное отличие между точками на диаграмме.
  На рисунке это видно как «раздвоение» ступенек (зелёная и красная траектории перестают совпадать).

- Это иллюстрирует **чувствительность к начальным условиям**:
  можно взять сколь угодно малую $\varepsilon$ (то есть сделать $y_0$ очень близким к $x_0$),

**Задание 5.** Бифуркационная диаграмма логистического отображения

Рассматриваем логистическое отображение
$$
x_{n+1}=f(x_n)=r\,x_n(1-x_n)
$$

**Идея бифуркационной диаграммы:** для каждого значения параметра $r$ итерационно строим последовательность
$\{x_n\}$, отбрасываем первые итерации, а затем наносим на график
несколько последних значений $x_n$ по вертикали напротив данного $r$.


```python
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'

def f(x, r):
    return r * x * (1 - x)

def bifurcation_data(r_min=0.0, r_max=4.0, n_r=4500,
                     n_iter=1800, n_trans=1200, keep=200, seed=42):
    """
    Возвращает сетку r и матрицу последних значений x
    после отбрасывания переходного процесса
    """
    rng = np.random.default_rng(seed)
    r = np.linspace(r_min, r_max, n_r)

    x = rng.random(n_r)

    saved = np.empty((keep, n_r))
    k = 0

    for i in range(n_iter):
        x = f(x, r)
        if i >= n_trans and k < keep:
            saved[k] = x
            k += 1

    return r, saved

r_vals, x_last = bifurcation_data()

r_inf = 3.5699456

plt.figure(figsize=(10, 6))

for k in range(x_last.shape[0]):
    plt.plot(r_vals, x_last[k], ',k', alpha=0.12)

plt.axvline(3.0, linestyle='--', alpha=0.6, label=r'$r=3$ (первая бифуркация)')
plt.axvline(r_inf, linestyle='--', alpha=0.6, label=rf'$r_\infty\approx {r_inf:.6f}$')

plt.title(r'Бифуркационная диаграмма логистического отображения $x_{n+1}=r x_n(1-x_n)$')
plt.xlabel(r'$r$')
plt.ylabel(r'$x_n$')
plt.xlim(0, 4)
plt.ylim(0, 1)
plt.grid(True, linestyle=':', alpha=0.5)
plt.legend()
plt.show()
```

## Интерпретация полученного графика

### 1) Как читать диаграмму
Для каждого $r$ по вертикали показаны значения $x_n$, которые остаются **после** того, как
переходный процесс закончился (мы отбросили первые итерации).  

- **Одна точка** при фиксированном $r$ → траектория стремится к **неподвижной точке**.
- **Две точки** → устойчивый **цикл периода 2**.
- **4, 8, 16 точек** → каскад **удвоения периода**.
- **Сплошная “полоса” точек** → **хаотический режим** (значения заполняют множество).

### 2) Что происходит при росте $r$ (основные режимы)
- При $0<r<1$ траектория стремится к $0$.
- При $1<r<3$ существует устойчивая неподвижная точка $x^*=1-\dfrac{1}{r}$ (на диаграмме это одна ветка).
- При $r=3$ происходит первая бифуркация: неподвижная точка теряет устойчивость и появляется цикл периода 2.
- При дальнейшем росте $r$ наблюдается каскад удвоений периода: $2\to4\to8\to\cdots$.

### 3) Где находится $r_\infty$ и поведение до/после него
На диаграмме $r_\infty$ — это **граница**, к которой “сдвигаются” точки бифуркаций удвоения периода.
Численно
$$
r_\infty \approx 3.5699456.
$$

- **До $r_\infty$**: устойчивые режимы периодические (сначала точка, потом циклы периодов $2^k$).
- **После $r_\infty$**: в основном начинается хаотическое поведение (много значений $x_n$ при одном $r$),
  но могут встречаться “окна периодичности” — интервалы $r$, где снова появляется устойчивый цикл.

**Вывод:** бифуркационная диаграмма показывает, как при изменении параметра $r$ система переходит
от устойчивой точки → к периодическим циклам → к хаосу.

**Задача 7.** Окна периодичности: найти периоды $3$, $5$, $6$ и отметить их на диаграмме

Идея: для каждого $r$ считаем траекторию после переходного процесса и проверяем,
становится ли хвост последовательности периодическим с периодом $p\in\{3,5,6\}$.
Затем отмечаем найденные интервалы $r$ на бифуркационной диаграмме.

```python
def detect_period_for_r(r, x0=0.217391, n_trans=3500, tail=900, pmax=30, tol=1e-7):
    x = x0
    for _ in range(n_trans):
        x = f(x, r)

    seq = np.empty(tail, dtype=float)
    for i in range(tail):
        x = f(x, r)
        seq[i] = x

    for p in range(1, pmax + 1):
        if 3 * p > tail:
            break
        block = seq[-3*p:]
        if np.max(np.abs(block[p:] - block[:-p])) < tol:
            return p
    return 0

def find_period_windows(r_min=3.4, r_max=4.0, n_r=4500, targets=(3,5,6),
                        n_trans=3500, tail=900, pmax=30, tol=1e-7):
    r_vals = np.linspace(r_min, r_max, n_r)
    periods = np.zeros(n_r, dtype=int)

    for i, r in enumerate(r_vals):
        periods[i] = detect_period_for_r(r, n_trans=n_trans, tail=tail, pmax=pmax, tol=tol)

    windows = {p: [] for p in targets}

    for p in targets:
        idx = np.where(periods == p)[0]
        if len(idx) == 0:
            continue
        start = idx[0]
        prev = idx[0]
        for j in idx[1:]:
            if j == prev + 1:
                prev = j
            else:
                windows[p].append((r_vals[start], r_vals[prev]))
                start = j
                prev = j
        windows[p].append((r_vals[start], r_vals[prev]))

    return r_vals, periods, windows

r_grid, periods, windows = find_period_windows()

min_width = 3e-4
for p in [3,5,6]:
    good = [(a,b) for (a,b) in windows[p] if (b-a) >= min_width]
    print(f'Период {p}: найдено окон = {len(good)}')
    for a, b in good[:8]:
        print(f'  r ∈ [{a:.6f}, {b:.6f}]   ширина ≈ {b-a:.6f}')
```

**Задача 8.** Устойчивость точки $x^*=0$ для $g(x)=r x(1-x^2)$

$$
x_{n+1}=g(x_n)=r\,x_n(1-x_n^2),\qquad r\in\left[0,\frac{3\sqrt3}{2}\right].
$$

### 1. Неподвижная точка
Точка $x^{*}=0$ является неподвижной, поскольку
$$
g(0)=r\cdot 0\cdot (1-0)=0.
$$


## 2. Устойчивость и асимптотическая устойчивость при $0\le r\le 1$

### 2.1. Устойчивость по определению
Докажем устойчивость точки $x^{*}=0$ при $0\le r\le 1$.

Пусть задано $\varepsilon>0$. Положим
$$
\delta=\min\{\varepsilon,1\}.
$$
Рассмотрим начальное значение $x_0$, удовлетворяющее $|x_0|<\delta$.
Покажем, что тогда $|x_n|<\varepsilon$ для всех $n\in\mathbb{N}$.

Пусть $|x_n|\le 1$. Тогда $0\le 1-x_n^{2}\le 1$, и поэтому
$$
|x_{n+1}|=|r\,x_n(1-x_n^{2})|
\le r|x_n|
\le |x_n|\qquad (\text{так как }0\le r\le 1).
$$
Следовательно,
$$
|x_n|\le |x_{n-1}|\le \dots \le |x_0|<\delta\le \varepsilon,\qquad \forall n.
$$
Таким образом, для любого $\varepsilon>0$ существует $\delta>0$, обеспечивающее попадание и удержание траектории в $\varepsilon$-окрестности нуля. Значит, $x^{*}=0$ **устойчива** при $0\le r\le 1$.

### 2.2. Асимптотическая устойчивость

**Случай $0\le r<1$.** Из оценки выше получаем
$$
|x_{n+1}|\le r|x_n|
\ \Longrightarrow\
|x_n|\le r^{\,n}|x_0|\xrightarrow[n\to\infty]{}0.
$$
Следовательно, $x_n\to 0$, т.е. точка $x^{*}=0$ **асимптотически устойчива** при $0\le r<1$.

**Случай $r=1$.** Пусть $x_0\in(0,1)$. Тогда $x_n\in(0,1)$ для всех $n$ и
$$
x_{n+1}=x_n(1-x_n^{2})<x_n,
$$
то есть последовательность $\{x_n\}$ монотонно убывает и ограничена снизу нулём, поэтому имеет предел $L=\lim_{n\to\infty}x_n\ge 0$.
Переходя к пределу в рекуррентном соотношении, получаем
$$
L=L(1-L^{2})
\ \Longrightarrow\
L^{3}=0
\ \Longrightarrow\
L=0.
$$
Следовательно, при $r=1$ также $x_n\to 0$, т.е. точка $x^{*}=0$ **асимптотически устойчива** (при начальных данных из $(0,1)$).


## 3. Неустойчивость при $r>1$
Докажем, что при $r>1$ точка $x^{*}=0$ неустойчива по определению.

Выберем произвольное $\varepsilon\in(0,1)$. Зафиксируем $\delta\in(0,\varepsilon)$ так, чтобы
$$
q=r(1-\delta^{2})>1,
$$
что возможно, поскольку $r>1$ и $1-\delta^{2}\to 1$ при $\delta\to 0$.

Тогда для любого $x$ с $0<|x|<\delta$ имеем $1-x^{2}\ge 1-\delta^{2}$ и, следовательно,
$$
|g(x)|=r|x|(1-x^{2})\ge r|x|(1-\delta^{2})=q|x|.
$$

Пусть задано произвольное $\delta_0>0$. Возьмём
$$
x_0=\min\left\{\frac{\delta_0}{2},\frac{\delta}{2}\right\},
$$
тогда $|x_0|<\delta_0$ и $|x_0|<\delta$.

Пока $|x_n|<\delta$, выполняется $|x_{n+1}|\ge q|x_n|$, откуда
$$
|x_n|\ge q^{\,n}|x_0|.
$$
Так как $q>1$, существует $n$ такое, что $q^{\,n}|x_0|>\varepsilon$, то есть траектория выходит из $\varepsilon$-окрестности нуля. Следовательно, условие устойчивости нарушается для любого $\delta_0>0$.

Значит, при $r>1$ точка $x^{*}=0$ **неустойчива**.


## 4. Итог
- $x^{*}=0$ устойчива при $0\le r\le 1$;
- $x^{*}=0$ асимптотически устойчива при $0\le r<1$ (а также при $r=1$ для $x_0\in(0,1)$);
- $x^{*}=0$ неустойчива при $r>1$.

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'

def g(x, r):
    return r * x * (1 - x**2)

def iterate_map(x0, r, n=250):
    x = x0
    for _ in range(n):
        x = g(x, r)
    return x

test_r = [0.5, 0.9, 0.99, 1.0, 1.01, 1.2, 1.5]
x0 = 0.2

print("Численная проверка около x*=0 (старт x0=0.2):\n")
for r in test_r:
    xN = iterate_map(x0, r, n=300)
    status = "≈ 0" if abs(xN) < 1e-2 else ""
    print(f"r={r:>4.2f}  ->  x_300={xN:.6f}  ({status})")

print("\nОжидаемая граница: r = 1")
```

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'

def g_map(x, r):
    return r * x * (1 - x**2)

def logistic_map(x, r):
    return r * x * (1 - x)

def bifurcation_points(map_func, r_min, r_max, n_r=2500, n_trans=1400, keep=140, x0s=(0.2, 0.7)):
    r_vals = np.linspace(r_min, r_max, n_r)
    rr, xx = [], []
    for r in r_vals:
        for x0 in x0s:
            x = x0
            for _ in range(n_trans):
                x = map_func(x, r)
            for _ in range(keep):
                x = map_func(x, r)
                rr.append(r)
                xx.append(x)
    return np.array(rr), np.array(xx)

r_max_g = 3*np.sqrt(3)/2
rr_g, xx_g = bifurcation_points(g_map, 0.0, r_max_g)

plt.figure(figsize=(10, 6))
plt.plot(rr_g, xx_g, ',k', alpha=0.12)
plt.title(r'Бифуркационная диаграмма для $g(x)=r\,x(1-x^2)$', fontsize=14)
plt.xlabel(r'$r$')
plt.ylabel(r'$x_n$')
plt.xlim(0.0, r_max_g)
plt.ylim(-0.02, 1.02)
plt.grid(True, linestyle=':', alpha=0.5)
plt.show()

rr_l, xx_l = bifurcation_points(logistic_map, 0.0, 4.0, n_r=3200, n_trans=1200, keep=120, x0s=(0.2, 0.4, 0.7))

plt.figure(figsize=(10, 6))
plt.plot(rr_l, xx_l, ',k', alpha=0.12)
plt.title(r'Бифуркационная диаграмма для $f(x)=r\,x(1-x)$', fontsize=14)
plt.xlabel(r'$r$')
plt.ylabel(r'$x_n$')
plt.xlim(0.0, 4.0)
plt.ylim(-0.02, 1.02)
plt.grid(True, linestyle=':', alpha=0.5)
plt.show()
```

## Сравнение с логистическим отображением

### Сходства
1. **Есть смена установившегося режима при росте $r$**: при малых $r$ наблюдается простое поведение (сходимость к неподвижной точке), а при увеличении параметра появляются более сложные режимы.
2. **Появляются области регулярности** (устойчивые циклы) и области более сложной динамики — на диаграмме это видно как ветвления и расслоение точек по $x_n$.
3. **Одинаковая граница устойчивости $x^{*}$= 0**: $r$ = 1

### Различия
1. **Интервал параметра:** для $g$ параметр ограничен сверху $r\le \frac{3\sqrt3}{2}$ (карта устроена так, чтобы траектории не “вылетали” из допустимой области), а для логистической карты - $r\in[0,4]$.
2. **Первая бифуркация при разных r:** у $f(x)=r\,x(1-x^2)$ первая бифуркация происходит при r = 2, у $f(x)=r\,x(1-x)$ первая бифуркация происходит при r = 3

**Вывод:** обе диаграммы демонстрируют усложнение динамики при росте $r$, в целом графики достаточно похожи, но есть отличия в деталях

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'

def g_map(x, r):
    return r * x * (1 - x**2)

def bifurcation_points(map_func, r_min, r_max, n_r=2200, n_trans=1400, keep=160, x0s=(0.2, 0.7)):
    r_vals = np.linspace(r_min, r_max, n_r)
    rr, xx = [], []
    for r in r_vals:
        for x0 in x0s:
            x = x0
            for _ in range(n_trans):
                x = map_func(x, r)
            for _ in range(keep):
                x = map_func(x, r)
                rr.append(r)
                xx.append(x)
    return np.array(rr), np.array(xx), r_vals

def detect_period_from_tail(tail, pmax=12, tol=1e-6):
    for p in range(1, pmax + 1):
        if 2*p > len(tail):
            break
        a = tail[-p:]
        b = tail[-2*p:-p]
        if np.max(np.abs(a - b)) < tol:
            return p
    return 0

def find_windows_for_periods(map_func, r_min, r_max, n_r=2000, n_trans=1800, tail_len=300, pmax=12, tol=1e-6, x0=0.217391):
    r_vals = np.linspace(r_min, r_max, n_r)
    periods = np.zeros(n_r, dtype=int)

    for i, r in enumerate(r_vals):
        x = x0
        for _ in range(n_trans):
            x = map_func(x, r)
        tail = np.empty(tail_len, dtype=float)
        for k in range(tail_len):
            x = map_func(x, r)
            tail[k] = x
        periods[i] = detect_period_from_tail(tail, pmax=pmax, tol=tol)

    windows = {}
    start = 0
    while start < n_r:
        p = periods[start]
        end = start + 1
        while end < n_r and periods[end] == p:
            end += 1
        if p > 0:
            a = r_vals[start]
            b = r_vals[end - 1]
            windows.setdefault(p, []).append((a, b))
        start = end

    return r_vals, periods, windows

r_min, r_max = 0.0, 3*np.sqrt(3)/2
r_grid, periods, windows = find_windows_for_periods(
    g_map, r_min, r_max,
    n_r=1800, n_trans=1600, tail_len=250,
    pmax=10, tol=2e-6, x0=0.23
)

rr, xx, _ = bifurcation_points(g_map, r_min, r_max, n_r=2400, n_trans=1400, keep=140)

plt.figure(figsize=(10, 6))
plt.plot(rr, xx, ',k', alpha=0.10)

plt.title(r'Окна периодичности на бифуркационной диаграмме для $g(x)=r\,x(1-x^2)$', fontsize=13)
plt.xlabel(r'$r$')
plt.ylabel(r'$x_n$')
plt.xlim(0.9, r_max)
plt.ylim(-0.02, 1.02)
plt.grid(True, linestyle=':', alpha=0.5)

palette = ['tab:blue','tab:orange','tab:green','tab:red','tab:purple','tab:brown','tab:pink','tab:olive']
period_list = sorted([p for p in windows.keys() if p > 0])

min_width = 5e-3

labels_done = set()
for idx, p in enumerate(period_list):
    color = palette[idx % len(palette)]
    for a, b in windows[p]:
        if (b - a) < min_width:
            continue
        lab = fr'период {p}' if p not in labels_done else None
        plt.axvspan(a, b, color=color, alpha=0.25, label=lab)
    labels_done.add(p)

plt.legend(fontsize=10, loc='upper left')
plt.show()
```
