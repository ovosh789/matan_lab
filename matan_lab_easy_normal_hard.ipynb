## EASY LEVEL

Рассматриваем одномерное логистическое отображение
$$
x_{n+1} = r x_n (1 - x_n), \quad r \in [0,1].
$$

**Задание 1.** Доказать, что если
$$
0 < x_0 < 1,
$$
то для всех $n \in \mathbb{N}$ выполняется
$$
0 < x_n < 1.
$$


### Инвариантность интервала $(0,1)$ для логистического отображения

Рассматривается отображение
$$
x_{n+1} = r x_n (1 - x_n), \quad r \in [0,1].
$$

Пусть начальное значение удовлетворяет
$$
0 < x_0 < 1.
$$
Покажем, что при всех шагах, когда $r > 0$, траектория остаётся внутри того же интервала:
$$
0 < x_n < 1 \quad \text{для всех } n \ge 1.
$$


## Шаг 1. Положительность следующего члена

Предположим, что на некотором шаге $n$ выполнено
$$
0 < x_n < 1.
$$
Тогда
$$
x_n > 0, \qquad 1 - x_n > 0.
$$
Если $r > 0$, то произведение трёх положительных множителей даёт
$$
x_{n+1} = r\,x_n(1 - x_n) > 0.
$$


## Шаг 2. Оценка сверху: $x_{n+1} < 1$

На отрезке $[0,1]$ максимум функции $x(1-x)$ равен $\dfrac{1}{4}$, поэтому
$$
0 < x_n(1 - x_n) \le \frac{1}{4}.
$$
Так как $r \le 1$, получаем
$$
x_{n+1} = r\,x_n(1 - x_n)
         \le 1 \cdot \frac{1}{4}
         = \frac{1}{4} < 1.
$$


## Шаг 3. Индукционный вывод

Так как исходное значение $x_0$ лежит в интервале $(0,1)$, а шаг «$n \to n+1$» сохраняет это свойство, по индукции имеем
$$
0 < x_n < 1 \quad \text{для всех } n \ge 0,\ \text{если } r \in (0,1].
$$


### Замечание о случае $ r = 0 $

При $r = 0$ отображение вырождается в
$$
x_{n+1} = 0.
$$
Тогда независимо от $x_0 \in (0,1)$ уже на первом шаге
$$
x_1 = 0 \notin (0,1).
$$
То есть при $r = 0$ интервал $(0,1)$ не является инвариантным, и утверждение выше верно именно для параметров $r \in (0,1]$.


### Влияние параметра \(r\) на поведение функции

Рассмотрим логистическое отображение
$$
x_{n+1} = r x_n (1 - x_n), \qquad 0 < x_n < 1.
$$

Поведение последовательности \(\{x_n\}\) зависит от значения параметра \(r\):

1. **Малые значения \(r\) ($0 < r \le 1$).**  
  Какое бы значение $x_0 \in (0,1)$ мы ни выбрали, каждое следующее значение меньше предыдущего, последовательность монотонно убывает.  
   Математически:
   $$
   \lim_{n \to \infty} x_n = 0.
   $$

2. **Средние значения \(r\) ($1 < r < 3$).**  
   Последовательность перестаёт уходить к нулю и со временем приближается к некоторому фиксированному числу.  
   Неподвижная точка:
   $$
   x^* = \frac{r - 1}{r},
   $$
   и при подходящих начальных условиях
   $$
   \lim_{n \to \infty} x_n = x^*.
   $$

3. **Большие значения \(r\) ($r \ge 3$).**  
   Последовательность в общем случае предела не имеет. При $r$ немного больше 3 появляются периодические циклы (значения повторяются с некоторым периодом), а при дальнейшем увеличении $r$ наблюдается сложное, хаотическое поведение.


```python
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'

def logistic_map(x, r):
    """Логистическое отображение x_{n+1} = r x_n (1 - x_n)."""
    return r * x * (1 - x)

# значения x_n из (0, 1)
x = np.linspace(0, 1, 400)

# три характерных значения параметра r
r_values = [0.6, 2.2, 3.5]
line_styles = ['-', '--', '-.']
labels = [
    r'$r = 0.6$',
    r'$r = 2.2$',
    r'$r = 3.5$'
]

plt.figure(figsize=(9, 6))

for r, ls, label in zip(r_values, line_styles, labels):
    plt.plot(x, logistic_map(x, r), ls, linewidth=2, label=label)

# диагональ y = x
plt.plot(x, x, color='gray', alpha=0.6, linewidth=1.5, label=r'$y = x$')

plt.title(r'Графики $x_{n+1}$ от $x_n$ для разных значений $r$', fontsize=14)
plt.xlabel(r'$x_n$', fontsize=12)
plt.ylabel(r'$x_{n+1}$', fontsize=12)
plt.grid(alpha=0.3)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.legend(fontsize=11)

plt.show()
```

### Сравнение логистического отображения и отображения варианта

Логистическое отображение:
$$
x_n = r x_{n-1} (1 - x_{n-1}).
$$

Отображение варианта при $N = 1$:
$$
x_n = r x_{n-1} (1 - x_{n-1}^2), \qquad
r \in \left[0, \dfrac{3\sqrt{3}}{2}\right].
$$

**Общие черты:**

- обе функции задают зависимость $x_n = f(x_{n-1})$ и содержат общий множитель $r x_{n-1}$;
- при увеличении параметра $r$ графики поднимаются вверх: для тех же $x_{n-1}$ значения $x_n$ становятся больше;
- пересечения с прямой $x_n = x_{n-1}$ во всех случаях соответствуют неподвижным точкам, их положение меняется при изменении $r$.

**Отличия:**

- в логистическом отображении нелинейный множитель равен $1 - x_{n-1}$ (получается парабола на $[0,1]$);
- в варианте множитель равен $1 - x_{n-1}^2$, из-за этого функция симметрична относительно точки $x_{n-1} = 0$ и имеет форму кубической кривой;
- для логистического отображения естественная область — $x_{n-1} \in [0,1]$, 
  у варианта поведение при отрицательных $x_{n-1}$ зеркально повторяет поведение при положительных;
- допустимый диапазон параметра $r$ в варианте ограничен отрезком $[0, \dfrac{3\sqrt{3}}{2}]$, поэтому значения $r$, при которых появляются разные структуры (сходимость, циклы, хаос), не совпадают с логистическим случаем.

**Вывод.**  
Сходство графиков объясняется тем, что обе модели имеют структуру
$$
x_n = r x_{n-1} \cdot \bigl(\text{нелинейный множитель от } x_{n-1}\bigr).
$$
Различия связаны с тем, что в логистическом отображении множитель зависит линейно от $x_{n-1}$, 
а в варианте — через квадрат $x_{n-1}^2$, что меняет симметрию графика и положение максимумов и неподвижных точек.


## NORMAL LEVEL ##

**Задание 1.**

Рассматриваем логистическое отображение
$$
x_{n+1} = f(x_n) = r x_n (1 - x_n).
$$

Неподвижная точка $x^*$ - это такое число, что
$$
x^* = f(x^*).
$$


### 1. Найти все неподвижные точки логистического отображения

Решаем уравнение
$$
x = r x (1 - x).
$$

Перенесём всё в одну сторону:
$$
r x (1 - x) - x = 0,
$$
$$
r x - r x^2 - x = 0,
$$
$$
-r x^2 + (r - 1)x = 0.
$$

Вынесем $x$ за скобки:
$$
x\bigl(-r x + (r - 1)\bigr) = 0.
$$

Отсюда два решения:
$$
x_1^* = 0,
$$
$$
-r x + (r - 1) = 0 \ \Rightarrow\ x_2^* = \frac{r - 1}{r} \quad (r \ne 0).
$$

Итак, **все неподвижные точки**:
- $x_1^* = 0$ при любом $r$;
- $x_2^* = \dfrac{r - 1}{r}$, если $r \ne 0$.

Если дополнительно считать, что $x \in [0,1]$ и $r \ge 0$, то $x_2^*$ попадает в $[0,1]$ только при $r \ge 1$.


### 2. При каких $r$ одна неподвижная точка, а при каких две?

С учётом области $x \in [0,1]$, $r \ge 0$:

- при $0 \le r < 1$ во внутренности отрезка есть только точка
  $$
  x^* = 0.
  $$
  Вторая формула даёт $x_2^* = \dfrac{r - 1}{r} < 0$, она вне отрезка $[0,1]$;

- при $r = 1$ снова получаем только $x^* = 0$ (второе решение совпадает с ним);

- при $r > 1$ обе точки лежат в $[0,1]$:
  $$
  x_1^* = 0,\qquad x_2^* = \frac{r - 1}{r} \in (0,1).
  $$

Итак:
- **одна** неподвижная точка в $[0,1]$ при $0 \le r \le 1$;
- **две** неподвижные точки в $[0,1]$ при $r > 1$.


### 3. Какое максимальное количество неподвижных точек может быть? Почему?

Неподвижные точки находятся из уравнения
$$
f(x) = x \quad \Longleftrightarrow \quad r x (1 - x) = x,
$$
которое мы свели к квадратному:
$$
-r x^2 + (r - 1)x = 0.
$$

Квадратное уравнение по $x$ может иметь не более **двух** различных корней.  
Значит, логистическое отображение не может иметь больше двух неподвижных точек.

Фактически:
- иногда корень один (когда $x_1^*$ и $x_2^*$ совпадают или второй лежит вне $[0,1]$);
- максимум - **две** неподвижные точки.


## NORMAL LEVEL ##

**Задание 2.**

### Доказательство монотонного убывания и сходимости

**Дано.** Логистическое отображение
$$
x_{n+1} = r x_n (1 - x_n),
$$
где начальное значение и параметр удовлетворяют
$$
0 < x_0 < 1, \qquad 0 < r \le 1.
$$


### 1. Монотонное убывание

Нужно показать, что для любого $n$ выполняется
$$
x_{n+1} \le x_n.
$$

Рассмотрим отношение следующего члена к текущему:
$$
\frac{x_{n+1}}{x_n}
= \frac{r x_n (1 - x_n)}{x_n}
= r(1 - x_n).
$$

Из инвариантности интервала $(0,1)$ знаем, что при $0 < x_0 < 1$ все последующие члены тоже лежат в этом интервале:
$$
0 < x_n < 1.
$$
Значит,
$$
0 < 1 - x_n < 1.
$$

Так как $0 < r \le 1$, имеем
$$
0 < r(1 - x_n) \le 1.
$$

Следовательно,
$$
\frac{x_{n+1}}{x_n} = r(1 - x_n) < 1
\Rightarrow x_{n+1} < x_n.
$$

Итак, последовательность $\{x_n\}$ **строго монотонно убывает**.


### 2. Существование и значение предела

Мы знаем:

1. $\{x_n\}$ монотонно убывает;
2. все члены последовательности положительны: $x_n > 0$ (произведение положительных множителей).

Значит, последовательность ограничена снизу нулём и монотонна, поэтому по теореме о монотонной ограниченной последовательности существует предел
$$
L = \lim_{n \to \infty} x_n.
$$

Найдём $L$, переходя к пределу в рекуррентной формуле:
$$
L = r L (1 - L).
$$

Перенесём всё в одну часть:
$$
L - rL(1 - L) = 0,
$$
$$
L(1 - r + rL) = 0.
$$

Получаем два возможных значения:

1. $L_1 = 0$;
2. $1 - r + rL = 0 \Rightarrow rL = r - 1 \Rightarrow L_2 = 1 - \dfrac{1}{r}$.

При $0 < r \le 1$ имеем $\dfrac{1}{r} \ge 1$, поэтому
$$
L_2 = 1 - \frac{1}{r} \le 0.
$$

Но все $x_n > 0$, значит предел не может быть отрицательным. Остаётся только
$$
L = 0.
$$

**Ответ.** При $0 < x_0 < 1$ и $0 < r \le 1$ последовательность $\{x_n\}$, заданная логистическим отображением, строго убывает, её предел существует и равен
$$
\lim_{n \to \infty} x_n = 0.
$$


```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'

def logistic_iter(x0, r, steps):
    """Итерации x_{n+1} = r x_n (1 - x_n)."""
    xs = [x0]
    x = x0
    for _ in range(steps):
        x = r * x * (1 - x)
        xs.append(x)
    return np.array(xs)

x0 = 0.8          # начальное значение
steps = 20        # число шагов
r_values = [0.2, 0.5, 1.0]

plt.figure(figsize=(8, 5))

for r in r_values:
    xs = logistic_iter(x0, r, steps)
    n = np.arange(len(xs))
    plt.plot(n, xs, "o-", linewidth=1.5, label=f"r = {r}")

# линия x = 0
plt.axhline(0, color="gray", linestyle="--", linewidth=1)

plt.title("Последовательность x_n при 0 < r <= 1")
plt.xlabel("n")
plt.ylabel("x_n")
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

**Дано:**

- логистическое отображение  
  $$x_{n+1} = r x_n (1 - x_n), \quad r \in (2,3);$$
- неподвижная точка  
  $$x^* = \frac{r-1}{r};$$
- условие расположения членов:  
  $$x_{2n} > x^*, \quad x_{2n+1} < x^*.$$

Функция
$$
f(x) = r x (1 - x)
$$
- парабола, обращённая вниз, с максимумом в точке $x = 0.5$.

Так как $r > 2$, то
$$
x^* = 1 - \frac{1}{r} > 1 - \frac{1}{2} = 0.5,
$$
значит неподвижная точка находится на **правой ветви** параболы, где функция $f(x)$ убывает.

Свойство убывающей функции: если $a > b$, то
$$
f(a) < f(b).
$$


### 1. Переход от чётного к нечётному

По условию $x_{2n} > x^*$. Применим к обеим частям неравенства функцию $f$.
Так как $f$ убывает на этом участке, знак меняется:
$$
f(x_{2n}) < f(x^*).
$$

Так как $x^*$ — неподвижная точка, $f(x^*) = x^*$, поэтому
$$
x_{2n+1} = f(x_{2n}) < x^*.
$$

То есть нечётные члены действительно лежат **слева** от $x^*$.


### 2. Переход от нечётного к чётному

Теперь из условия $x_{2n+1} < x^*$ опять применим убывающую функцию $f$:
$$
f(x_{2n+1}) > f(x^*) = x^*,
$$
следовательно,
$$
x_{2n+2} = f(x_{2n+1}) > x^*.
$$

Значит чётные члены снова оказываются **справа** от $x^*$.


### 3. Монотонность подпоследовательностей

Рассмотрим составную функцию
$$
g(x) = f(f(x)).
$$

Функция $f(x)$ убывает на рассматриваемом участке, поэтому $g(x)$ — «убывающая от убывающей»,
то есть **возрастающая**.

Из графика видно, что при $2 < r < 3$:

- для точек правее неподвижной точки
  $$x > x^* \quad \Rightarrow \quad g(x) < x;$$
- для точек левее неподвижной точки
  $$x < x^* \quad \Rightarrow \quad g(x) > x.$$

Теперь подставим наши подпоследовательности.

**Чётная подпоследовательность** $u_n = x_{2n}$.  
У нас $u_n > x^*$, значит
$$
u_{n+1} = x_{2n+2} = g(u_n) < u_n.
$$
Следовательно, $\{x_{2n}\}$ **монотонно убывает** и остаётся справа от $x^*$.

**Нечётная подпоследовательность** $v_n = x_{2n+1}$.  
Здесь $v_n < x^*$, поэтому
$$
v_{n+1} = x_{2n+3} = g(v_n) > v_n.
$$
Значит, $\{x_{2n+1}\}$ **монотонно возрастает** и остаётся слева от $x^*$.

**Итог:** при $2 < r < 3$ и условии $x_{2n} > x^*,\ x_{2n+1} < x^*$:

- чётная подпоследовательность $\{x_{2n}\}$ убывает к $x^*$ сверху;
- нечётная подпоследовательность $\{x_{2n+1}\}$ возрастает к $x^*$ снизу;
- обе подпоследовательности стремятся к одной и той же точке равновесия
  $$x^* = \dfrac{r-1}{r}.$$


```python
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'

def logistic(x, r):
    return r * x * (1 - x)

r = 2.9          # параметр из (2, 3)
x_star = (r - 1) / r
x0 = 0.8         # старт справа от x*
steps = 20

# считаем последовательность x_n
xs = [x0]
for _ in range(steps):
    xs.append(logistic(xs[-1], r))
xs = np.array(xs)
n = np.arange(len(xs))

# индексы чётных и нечётных членов
even_idx = n[::2]   # 0, 2, 4, ...
odd_idx  = n[1::2]  # 1, 3, 5, ...

plt.figure(figsize=(8, 5))

plt.plot(n, xs, "-.", color="gray", alpha=0.4, linewidth=1)

# чётные и нечётные отдельно
plt.scatter(even_idx, xs[even_idx], color="tab:blue", label="четные x_{2n}")
plt.scatter(odd_idx,  xs[odd_idx],  color="tab:orange", label="нечетные x_{2n+1}")

# горизонтальная линия x*
plt.axhline(x_star, color="green", linestyle="--", label="неподвижная точка x*")

plt.xlabel("номер шага n")
plt.ylabel("x_n")
plt.title(f"Четные и нечетные подпоследовательности (r = {r})")
plt.grid(alpha=0.3)
plt.legend(loc="upper right")

plt.show()
```

Рассматриваем отображение:

### 1. Аналитический поиск неподвижных точек

Неподвижная точка удовлетворяет уравнению $x = g(x)$:
$$
x = r x (1 - x^2).
$$

Перенесём всё в одну часть и вынесем $x$ за скобку:
$$
x - r x (1 - x^2) = 0,
$$
$$
x \bigl(1 - r(1 - x^2)\bigr) = 0.
$$

Отсюда два случая.

1. **Тривиальная точка**
   $$
   x_1^* = 0 \quad \text{(существует при любом } r > 0\text{)}.
   $$

2. **Нетривиальные точки** из уравнения
   $$
   1 - r(1 - x^2) = 0
   \;\Longrightarrow\;
   r(1 - x^2) = 1
   \;\Longrightarrow\;
   x^2 = 1 - \frac{1}{r}.
   $$
   Тогда
   $$
   x_{2,3}^* = \pm \sqrt{1 - \frac{1}{r}}.
   $$
   Они существуют, когда подкоренное выражение неотрицательно:
   $$
   1 - \frac{1}{r} \ge 0 \;\Longrightarrow\; r \ge 1.
   $$
   В интервале $0 < x < 1$ нас интересует только положительная точка
   $$
   x_+^* = \sqrt{1 - \frac{1}{r}}, \qquad r > 1.
   $$


### 2. Диапазон $r$, при котором $\{x_n\}$ монотонно сходится к нулю

Хотим, чтобы последовательность была положительной и строго убывающей:
$$
0 < x_{n+1} < x_n.
$$

Подставляем отображение:
$$
x_{n+1} = r x_n (1 - x_n^2) < x_n.
$$

Так как $x_n > 0$, делим неравенство на $x_n$:
$$
r(1 - x_n^2) < 1.
$$

Для $0 < x_n < 1$ выполнено $0 < 1 - x_n^2 < 1$. Если
$$
0 < r < 1,
$$
то
$$
0 < r(1 - x_n^2) \le r < 1,
$$
и неравенство $x_{n+1} < x_n$ выполняется для всех $n$.

При $r = 1$ имеем
$$
\frac{x_{n+1}}{x_n} = 1 - x_n^2 < 1 \quad (x_n > 0),
$$
то есть последовательность тоже строго убывает.

Кроме того, для $0 < r \le 1$ и $0 < x_n < 1$ выполняется
$$
x_{n+1} = r x_n (1 - x_n^2) > 0,
$$
а максимум функции $x(1 - x^2)$ на интервале $(0,1)$ равен
$$
\max_{x \in (0,1)} x(1 - x^2) = \frac{2}{3\sqrt{3}} < 1,
$$
поэтому
$$
x_{n+1} \le r \cdot \frac{2}{3\sqrt{3}} < 1.
$$
Значит все члены $\{x_n\}$ остаются в интервале $(0,1)$.

Таким образом, при
$$
0 < r \le 1
$$
последовательность $\{x_n\}$ положительна, ограничена и строго убывает.
Следовательно, существует предел
$$
L = \lim_{n\to\infty} x_n.
$$

Переходим к пределу в рекуррентной формуле:
$$
L = r L (1 - L^2).
$$

Получаем
$$
L \bigl(1 - r(1 - L^2)\bigr) = 0,
$$
откуда
$$
L = 0 \quad \text{или} \quad L^2 = 1 - \frac{1}{r}.
$$

При $0 < r \le 1$ величина $1 - \dfrac{1}{r} \le 0$, поэтому второе равенство
не даёт положительного предела (а все $x_n > 0$). Остаётся единственный
возможный предел:
$$
L = 0.
$$

**Итог:**

Неподвижные точки:
  $$
  x_1^* = 0 \quad (r > 0), \qquad
  x_{2,3}^* = \pm \sqrt{1 - \dfrac{1}{r}} \quad (r \ge 1).
  $$
- При $0 < r \le 1$ и $0 < x_0 < 1$ последовательность $\{x_n\}$
  **монотонно убывает и сходится к нулю**:
  $$
  \lim_{n\to\infty} x_n = 0.
  $$

```python
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'

def g(x, r):
    return r * x * (1 - x**2)

# Параметры
n_steps = 40
x0 = 0.6   # начальное значение в (0, 1)

r_values = [0.5, 1.4, 2.3]

labels = [
    "r = 0.5 (сходимость к 0)",
    "r = 1.4 (сходимость к x*>0)",
    "r = 2.3 (колебания / период - 2)"
]
colors = ["#27ae60", "#2980b9", "#c0392b"]

plt.figure(figsize=(10, 6))

for r, label, color in zip(r_values, labels, colors):
    x = x0
    path = [x]
    for _ in range(n_steps):
        x = g(x, r)
        path.append(x)

    n = np.arange(len(path))
    plt.plot(n, path, "o-", label=label, color=color, linewidth=2, markersize=4)

plt.title("Зависимость $x_n$ от $n$ для $g(x) = r x (1 - x^2)$", fontsize=14)
plt.xlabel("$n$", fontsize=12)
plt.ylabel("$x_n$", fontsize=12)
plt.axhline(0, color="black", linestyle="--", alpha=0.5)
plt.grid(True, linestyle=":", alpha=0.6)
plt.legend(fontsize=11)
plt.ylim(-0.05, 1.05)

plt.show()
```

### Задание 1 (изменение порядка цикла в зависимости от r, анализ значений m)

```python
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'

def logistic(x, r):
    return r * x * (1 - x)

def find_cycle_length_general(r, x0=0.5, max_period=1000, tol=1e-10, transient=2000, record=200):
    """
    Пытается найти наименьший период m (1 <= m <= max_period),
    такой что |f^m(x) - x| < tol после транзиента
    
    Возвращает:
        m — длина цикла (любое целое число), либо None
    """
    x = x0
    # Прогрев (транзиент)
    for _ in range(transient):
        x = logistic(x, r)
    
    x_start = x
    
    # Проверяем все m от 1 до max_period
    for m in range(1, max_period + 1):
        x_test = x_start
        for _ in range(m):
            x_test = logistic(x_test, r)
        if abs(x_test - x_start) < tol:
            return m
    return None  # не найден цикл длиной ≤ max_period


r_min, r_max = 3.0, 3.5699456
r_vals = np.linspace(r_min, r_max, 400)

cycle_lengths = []
for r in r_vals:
    m = find_cycle_length_general(r, x0=0.6, max_period=1000, tol=1e-10, transient=2000)
    cycle_lengths.append(m if m is not None else np.nan)  # используем nan для отсутствующих

# Визуализация
plt.figure(figsize=(12, 5))
plt.plot(r_vals, cycle_lengths, 'bo', markersize=3, alpha=0.8)
plt.yscale('log', base=2)
powers_of_2 = [1, 2, 4, 8, 16, 32, 64, 1024]
plt.yticks(powers_of_2, [str(p) for p in powers_of_2])
plt.xlabel('$r$')
plt.ylabel('Обнаруженная длина цикла $m$')
plt.title('Экспериментальное определение длины цикла при $r \\in (3,\\ r_\\infty)$')
plt.grid(True, which='both', linestyle='--', alpha=0.6)
plt.axvline(3.0, color='red', linestyle=':', label='$r = 3$')
plt.axvline(r_max, color='purple', linestyle=':', label='$r_\\infty \\approx 3.5699$')
plt.legend()
plt.ylim(0.9, 1000)
plt.show()



```

```python
r_zoom = np.linspace(3.568, 3.57, 100)
cycle_lengths_zoom = []
for r in r_zoom:
    m = find_cycle_length_general(r, x0=0.6, max_period=1000, tol=1e-10, transient=2000)
    cycle_lengths_zoom.append(m if m is not None else np.nan)

plt.figure(figsize=(12, 5))
plt.plot(r_zoom, cycle_lengths_zoom, 'bo', markersize=3, alpha=0.8)
plt.yscale('log', base=2)
plt.yticks([2, 4, 8, 16, 32, 64], ['2','4','8','16','32','64'])
plt.xlabel('$r$')
plt.ylabel('Обнаруженная длина цикла $m$')
plt.title('Увеличенный фрагмент: $r \\in [3.568,\\ 3.57]$')
plt.grid(True, linestyle='--', alpha=0.6)
plt.axvline(3.5687, color='gray', linestyle=':', label='$r_{32}$')
plt.axvline(3.5696, color='gray', linestyle=':', label='$r_{64}$')
plt.legend()
plt.ylim(1, 100)
plt.show()
```

На графиках видно чёткое удвоение периода:
    При r ≈ 3.0 - появляется цикл порядка 2,

### Задание 2 (Лестница Ламерея, циклы различных порядков на графике)

```python
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'

def logistic(x, r):
    """Логистическое отображение"""
    return r * x * (1 - x)

def cobweb_plot(f, x0, r, n_iter=50, x_range=(0, 1), title=""):
    """
    Строит лестницу Ламерея для отображения f при параметре r
    
    Параметры:
        f: функция отображения (x, r) -> x_next
        x0: начальное значение
        r: управляющий параметр
        n_iter: количество итераций для построения лестницы
        x_range: диапазон значений по осям
        title: заголовок графика
    """
    # Генерируем сетку для построения графика функции
    x = np.linspace(x_range[0], x_range[1], 400)
    y = f(x, r)
    
    # Инициализируем траекторию
    xn = [x0]
    yn = [0]  # старт с оси X
    x_curr = x0
    
    # Генерируем точки лестницы
    for i in range(n_iter):
        # Вертикаль: от (x_curr, x_curr) до (x_curr, f(x_curr))
        x_next = f(x_curr, r)
        xn.append(x_curr)
        yn.append(x_next)
        
        # Горизонталь: от (x_curr, x_next) до (x_next, x_next)
        xn.append(x_next)
        yn.append(x_next)
        
        # Обновляем текущее значение для следующей итерации
        x_curr = x_next
    
    # Построение графика
    plt.figure(figsize=(8, 8))
    plt.plot(x, y, 'b', lw=2, label='$f(x)$')
    plt.plot(x, x, 'k--', lw=1, label='$y = x$')
    plt.plot(xn, yn, 'r-', lw=1, alpha=0.8, label='Лестница Ламерея')
    
    plt.xlim(x_range)
    plt.ylim(x_range)
    plt.xlabel('$x_n$')
    plt.ylabel('$x_{n+1}$')
    plt.title(title if title else f'Лестница Ламерея: $r = {r}$')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
```

```python
# При r = 2.5 - неподвижная точка устойчива
cobweb_plot(logistic, x0=0.2, r=2.5, n_iter=30, title=r'Цикл периода 1 ($r = 2.5$)')
```

Лестница попадает в неподвижную точку, то есть образует цикл порядка 1

```python
# При r = 3.2 - цикл периода 2
cobweb_plot(logistic, x0=0.2, r=3.2, n_iter=50, title=r'Цикл периода 2 ($r = 3.2$)')
```

Лестница образует прямоугольник, переходя между двумя точками, что является подтверждением существования устойчивого цикла порядка 2

```python
# При r = 3.5 - цикл периода 4
cobweb_plot(logistic, x0=0.3, r=3.5, n_iter=100, title=r'Цикл периода 4 ($r = 3.5$)')
```

Лестница обходит четыре точки, создавая замкнутый путь, что указывает на наличие цикла порядка 4

```python
# При r = 4 - нет цикла
cobweb_plot(logistic, x0=0.3, r=4.0, n_iter=100, title=r'$r = 4.0$')
```

Лестница не замыкается, то есть цикла нет

### Задание 3 (анализ порядка цикла в зависимости от r для g(x)

```python
%config InlineBackend.figure_format = 'retina'

def g(x, r):
    return r * x * (1 - x**2)

r_max = 3 * np.sqrt(3) / 2


cobweb_plot(g, x0=0.4, r=1.0, n_iter=30, title=r'$r = 1.0$')

cobweb_plot(g, x0=0.1, r=1.2, n_iter=100, title=r'$r = 1.2$')

cobweb_plot(g, x0=0.3, r=2.1, n_iter=100, title=r'$r = 2.1$')

cobweb_plot(g, x0=0.3, r=2.25, n_iter=100, title=r'$r = 2.25$')

cobweb_plot(g, x0=0.1, r=r_max, n_iter=100, title=r'$r = r_{\max}$')
```

Мы видим, что при r = 1.0 и r = 1.2 образуется цикл порядка 1, при r = 2.1 - цикл порядка 2, при r = 2.25 - цикл порядка 4, а при r = r_max цикла нет (хаос).
